```python
# Exotic Pet Store

# my class of inventory is going to have a list of stock
class Inventory:
    # don't need to pass anything to initialize
    def __init__(self):
        # animal object and how many we have
        self.stock = {}
    
    # fill up my inventory
    def restock(self):
        fox = Animal('Fennec Fox', 4500.0)
        wolf = Animal('Grey Wolf', 3200.0)
        llama = Animal('Llama', 1500.0)
        alpaca = Animal('Alpaca', 2000.0)
        capybara = Animal('Capybara', 3400.0)
        self.addtostock(fox, 8)
        self.addtostock(wolf, 1)
        self.addtostock(llama, 4)
        self.addtostock(alpaca, 1)
        self.addtostock(capybara, 1)
    
    # adding a new animal to my stock
    def addtostock(self, a, q):
        self.stock[a.name.lower()] = {
            'obj': a,
            'quantity': q
        }
    
    def show(self):
        print("Welcome to Sam's Exotic Pets - here's what we have for sale today!")
        for i in self.stock:
            print(f"{self.stock[i]['quantity']} of {self.stock[i]['obj'].name} @ ${self.stock[i]['obj'].price:.2f} ea.")
        print('----------------')
        
    
    # remove an animal from my stock
    def removefromstock(self, item):
        if item in self.stock:
            del self.stock[item]
        else:
            return -1

#my inventory will be full of animal objects
class Animal:
    def __init__(self, n, p):
        self.name = n
        self.price = p
    
    def show(self):
        print(f'This is a {self.name} - it costs ${self.price:.2f}')

        
class Cart:
    # what is this person planning to buy
    def __init__(self):
        self.cart = {}
        self.total = 0.0
    
    # check if the item is in my inventory, if it is, add it to the cart
    def add(self, item, inv):
        if item in inv.stock:
            if item in self.cart:
                self.cart[item]['quantity'] += 1
                self.total += inv.stock[item]['obj'].price
            else:
                self.cart[item.lower()] = {
                    'obj': inv.stock[item]['obj'],
                    'quantity': 1
                }
                self.total += inv.stock[item]['obj'].price
            if inv.stock[item]['quantity'] == 1:
                del inv.stock[item]
            else:
                inv.stock[item]['quantity'] -= 1
            print(f'{item.title()} has been added to your cart!')
        else:
            print(f"We don't have {item.title()} in stock at the moment.")
    
    def show(self):
        for i in self.cart:
            print(f"{self.cart[i]['quantity']} of {self.cart[i]['obj'].name} @ ${self.cart[i]['obj'].price:.2f} ea.")
        print('=======')
        print(f'Total: ${self.total:.2f}')
    
    def delete(self, item, inv):
        if item in self.cart:
            if item in inv.stock:
                inv.stock[item]['quantity'] += 1
            else:
                inv.stock[item] = self.cart[item]
                inv.stock[item]['quantity'] = 1
            if self.cart[item]['quantity'] == 1:
                del self.cart[item]
            else:
                self.cart[item]['quantity'] -= 1
            self.total -= inv.stock[item]['obj'].price
            print(f'{item.title()} has been removed from your cart!')
        else:
            print(f"You don't have {item.title()} in your cart at the moment.")
    
    def checkout(self):
        self.show()
        print(f"Thanks for shopping at Sam's Exotic Pets!")

class Main:
    @classmethod
    def run(self):
        inv = Inventory()
        inv.restock()
        cart = Cart()
        flag = True
        print('''
            Welcome to Sam's Exotic Pets!
            'Buy' to buy something
            'View cart' to see your current cart
            'View animals' to see our current offerings
            'Remove' to put an animal back
            'Checkout' to checkout and leave!
            ''')
        while flag:
            ui = input('What would you like to do? ').lower()
            print('\n')
            if ui == 'view cart':
                cart.show()
            elif ui == 'view animals':
                inv.show()
            elif ui == 'remove':
                ur = input('What would you like to remove from your cart? ').lower()
                cart.delete(ur, inv)
            elif ui == 'buy':
                ub = input('Which animal would you like to buy? ').lower()
                cart.add(ub, inv)
            elif ui == 'checkout':
                flag = False
                cart.checkout()
            else:
                print('Invalid input, please try again.')
    
```

```python
Main.run()
```

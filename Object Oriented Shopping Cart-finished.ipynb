```python
from IPython.display import clear_output


class Cart:
    items = []
    
    def add_item(self, item):
        return self.items.append(item)

    def remove_item(self, item_str):
        for obj in self.items:
            if item_str == obj.name:
                self.items.remove(obj)
                break


    def show_items(self):
        total = 0
        print('=' * 40)
        if not self.items:
            print("You have no items in your cart. ")
        else:
            display_cart = []
            for p in self.items:
                if p.name not in [i['name'] for i in display_cart]:
                    display_cart.append({
                        'name': p.name,
                        'price': p.price
                    })
                else:
                    for i in display_cart:
                        if i['name'] == p.name:
                            i['price'] += p.price

            for i in range(len(display_cart)):
                print(f"{i+1}: {display_cart[i]['name']} [{[i.name for i in self.items].count(display_cart[i]['name'])}] ${display_cart[i]['price']}")

        print('=' * 40)

        total += float(sum([p.price for p in self.items]))

        print(f"Total: ${total:,.2f}")

    def clear_items(self):
        self.items.clear()
    
class Item:
    def __init__(self, name, price):
        self.name = name
        self.price = price
        
class ShoppingCart:
    def show_instructions():
        print("""Type 'add' to add an item to your cart.
Type 'remove' to remove an item from your cart.
Type 'clear' to remove all items from your cart.
Type 'show' to show your cart items.
Type 'quit' to exit program and show your cart.
""")
    
    @classmethod
    def run(self):
        
        cart = Cart()

        done = False
        while not done:
            clear_output()
            self.show_instructions()
            cart.show_items()
            
            ask = input("What would you like to do? ").lower()

            if ask == 'add':
                item_name = input("What item would you like to add? ")
                item_price = float(input("How much does the item cost? "))
                cart_item = Item(item_name, item_price)
                cart.add_item(cart_item)
            elif ask == 'remove':
                cart_item = input("What item would you like to remove? ")
                cart.remove_item(cart_item)
            elif ask == 'clear':
                confirm = input("Are you sure? This action cannot be undone. Y/N ").lower()
                if confirm == 'y':
                    cart.clear_items()
                elif confirm ==  'n':
                    continue
                else:
                    print("Invalid command, try again!")
                input("Press 'ENTER' to continue")
            elif ask == 'quit':
                confirmed = False

                while not confirmed:
                    confirm = input("Are you sure? Y/N ").lower()
                    if confirm == 'y':
                        done = True
                        confirmed = True
                    elif confirm == 'n':
                        break
                    else:
                        print("Invalid command, try again!")
                        input("Press 'ENTER' to continue")
        clear_output()
        cart.show_items()
        print("Have a good day! Goodbye!")
        
ShoppingCart.run()
```
